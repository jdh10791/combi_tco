```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

### Simple linear regression for 1 parameter

```python
x = np.arange(0,100,1)
true_y = x*2
noise = np.random.normal(size=100)*10
obs_y = true_y + noise
plt.scatter(x,obs_y,s=2,label='Noisy data',c='r')
plt.plot(x,true_y,label='Actual values')
plt.legend()
```

```python
a = np.vstack([x,np.zeros(len(x))]).T
sol = np.linalg.lstsq(a,obs_y,rcond=None)

fit = np.polyval(sol[0],x)
plt.plot(x,fit,label='Model',c='g')
plt.scatter(x,obs_y,s=2,c='r',label='Noisy data')
plt.legend()
```

### Multivariable linear regression for 2 parameters

```python
x = np.random.normal(size=(100,2))
true_y = x[:,0]*5 - x[:,1]*2
noise = np.random.normal(size=100)*2
obs_y = true_y + noise

#plot in 3d
fig = plt.figure()
#meshgrid for plotting actual plane
xx, yy = np.meshgrid(range(-4,4),range(-4,4))
z = xx*5 - yy*2
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(xx,yy,z,alpha=0.3)#,label='Actual function',color='b')
ax.scatter(x[:,0],x[:,1],obs_y,c='r',label='Noisy data')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.legend()
```

```python
sol2 = np.linalg.lstsq(x,obs_y,rcond=None)[0]
fit2 = xx*sol2[0] + yy*sol2[1]

#plot in 3d
fig = plt.figure()
#meshgrid for plotting actual plane
xx, yy = np.meshgrid(range(-4,4),range(-4,4))
z = xx*5 - yy*2
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(xx,yy,z,alpha=0.3)#,label='Actual function',color='b')
ax.plot_surface(xx,yy,fit2,alpha=0.5)
#ax.scatter(x[:,0],x[:,1],obs_y,c='r',label='Noisy data')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.legend()
```
