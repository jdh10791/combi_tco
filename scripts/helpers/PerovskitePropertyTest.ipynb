```python
#check feature labels
p = Perovskite('BaCo0.1Fe0.4Zr0.5O3',
               cation_site={'Ba':'A','Co':'B','Fe':'B','Zr':'B','Y':'B'},
               site_ox_lim={'A':[2,4],'B':[2,4]},
               site_base_ox={'A':2,'B':4,'X':-2},
               silent=True)

ox_stats= ['min','max','median']

for sites in ['A','B','X','ABX','','AB','AX','BX',['comp'],['A','B','comp']]:
    features = p.featurize(sites=sites,ox_stats=ox_stats)
    feature_labels = p.feature_labels(sites=sites,ox_stats=ox_stats)
    eq = (feature_labels==list(features.index))
    print(f'feature_labels() matches features.index for sites {sites}: {eq}')
```
